```python
import pandas as pd
```

```python
df = pd.read_csv("vine_table.csv")
```

```python
#df where total votes greater than 20
total_twenty_df = df[df["total_votes"] > 20]
total_twenty_df.head()
```

```python
#total_twenty_df where "helpful_votes" / "total_votes" is greater than 50%
helpful_total_df = total_twenty_df[(total_twenty_df["helpful_votes"] / total_twenty_df["total_votes"]) >= 0.5]
helpful_total_df.head()
```

```python
#helpful_total_df where review was written as part of the vine program
vine_y_df = helpful_total_df[helpful_total_df["vine"] == "Y"]
vine_y_df.head()
```

```python
#helpful_total_df where review was not written as part of the vine program
vine_n_df = helpful_total_df[helpful_total_df["vine"] == "N"]
vine_n_df.head()
```

```python
#total number of reviews (using df)
total_reviews = df['review_id'].count()
#total number of 5 star reviews
total_five_star = (df['review_id'][df['star_rating'] == 5]).count()
#total number of 5 star reviews that were paid
total_five_star_paid = (df['review_id'][(df['star_rating'] == 5) & (df['vine'] == 'Y')]).count()
#total number of 5 star reviews that were not paid
total_five_star_not_paid = (df['review_id'][(df['star_rating'] == 5) & (df['vine'] == 'N')]).count()
```
